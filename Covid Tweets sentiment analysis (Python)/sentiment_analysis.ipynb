```python
import pandas as pd
import numpy as np
from keras.callbacks import EarlyStopping
from keras.models import Sequential
from keras.layers import Embedding
from keras.layers import Dense, Input
from keras.layers import LSTM
from keras.layers import Dropout
from keras.layers import Bidirectional
from keras.layers import GlobalMaxPool1D,GlobalAveragePooling1D
import nltk
from nltk.stem import WordNetLemmatizer
import re
from nltk.corpus import stopwords
from keras.preprocessing.text import one_hot
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
import gensim.downloader as api
import tensorflow as tf
from sklearn.metrics import precision_score,recall_score,f1_score,accuracy_score

nltk.download('wordnet')
nltk.download('stopwords')
lem=WordNetLemmatizer()
```

```python
df=pd.read_csv('/content/Corona_NLP_train.csv',encoding = "ISO-8859-1")
#df.dropna(axis=0,inplace=True)
df
```

```python
Y=df.iloc[:,5]
Y=pd.get_dummies(Y)
Y
```

```python
X=df['OriginalTweet']
X=np.array(X)
X
```

```python


for i,sent in enumerate(X):


  sent=sent.lower()
  sent=re.sub('https:\/\/\S+','',sent)
  sent=re.sub('@\S+','',sent)
  sent=re.sub('&[a-zA-Z]+;','',sent)
  sent=re.sub('#\S+','',sent)
  sent=re.sub('[^a-zA-Z0-9]',' ',sent)
  sent=sent.split()
  wds=''
  for word in sent:
    if word not in stopwords.words('english'):
      word=lem.lemmatize(word)
      wds+=' '
      wds+=word
  X[i]=wds
X
```

```python

tokenizer=Tokenizer(num_words=30000)
tokenizer.fit_on_texts(X)
dic=tokenizer.word_index
print(dic)
```

```python
X_train=tokenizer.texts_to_sequences(X)
X_train
```

```python

X_train=pad_sequences(X_train,maxlen=500,padding='pre')
X_train
```

```python
model=Sequential()
model.add(Embedding(30000,300))
model.add(LSTM(500))
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(1024, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(5,activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
earlystopping = EarlyStopping(monitor='val_accuracy', patience=5, verbose=1)
```

```python
model.fit(X_train,Y,validation_split=.2,epochs=20,callbacks=[earlystopping],batch_size=20,verbose=1)
```

## **BERT**

```python
import tensorflow_hub as hub
!wget --quiet https://raw.githubusercontent.com/tensorflow/models/master/official/nlp/bert/tokenization.py
m_url = 'https://tfhub.dev/tensorflow/bert_en_uncased_L-12_H-768_A-12/2'
bert_layer = hub.KerasLayer(m_url, trainable=True)
```

```python
!pip install sentencepiece
import tokenization
```

```python
vocab_file = bert_layer.resolved_object.vocab_file.asset_path.numpy()
do_lower_case = bert_layer.resolved_object.do_lower_case.numpy()
bert_tokenizer = tokenization.FullTokenizer(vocab_file, do_lower_case)

def bert_encode(texts, tokenizer, max_len=512):
    all_tokens = []
    all_masks = []
    all_segments = []
    
    for text in texts:
        text = bert_tokenizer.tokenize(text)
        
        text = text[:max_len-2]
        input_sequence = ["[CLS]"] + text + ["[SEP]"]
        pad_len = max_len-len(input_sequence)
        
        tokens = bert_tokenizer.convert_tokens_to_ids(input_sequence) + [0] * pad_len
        pad_masks = [1] * len(input_sequence) + [0] * pad_len
        segment_ids = [0] * max_len
        
        all_tokens.append(tokens)
        all_masks.append(pad_masks)
        all_segments.append(segment_ids)
        
    return np.array(all_tokens), np.array(all_masks), np.array(all_segments)
```

```python
def build_model(bert_layer, max_len=512):
    input_word_ids = Input(shape=(max_len,), dtype=tf.int32, name="input_word_ids")
    input_mask = Input(shape=(max_len,), dtype=tf.int32, name="input_mask")
    segment_ids = Input(shape=(max_len,), dtype=tf.int32, name="segment_ids")
    
    pooled_output, sequence_output = bert_layer([input_word_ids, input_mask, segment_ids])
    
    clf_output = sequence_output[:, 0, :]
    
    lay = Dense(64, activation='relu')(clf_output)
    lay = Dropout(0.2)(lay)
    lay = Dense(32, activation='relu')(lay)
    lay = Dropout(0.2)(lay)
    out = Dense(5, activation='softmax')(lay)
    
    model = tf.keras.models.Model(inputs=[input_word_ids, input_mask, segment_ids], outputs=out)
    model.compile(tf.keras.optimizers.Adam(lr=2e-5), loss='categorical_crossentropy', metrics=['accuracy'])
    
    return model
```

```python
max_len = 250
train_input = bert_encode(df.OriginalTweet.values, tokenizer, max_len=max_len)
```

```python
modelbert = build_model(bert_layer, max_len=max_len)
```

```python
earlystopping = EarlyStopping(monitor='val_accuracy', patience=3, verbose=1)
modelbert.fit(train_input, Y, validation_split=0.2, epochs=10, callbacks=[earlystopping], batch_size=20, verbose=1)
```
