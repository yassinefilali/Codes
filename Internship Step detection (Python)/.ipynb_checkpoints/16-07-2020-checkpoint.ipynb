## Covariance with cumulative sum

$$C_{u,v}=\frac{1}{v-u}\sum_{i=u}^{v-1}(y_i-\bar{y}_{u,v})(y_i-\bar{y}_{u,v})^T$$  
$$=\frac{1}{v-u}\sum_{i=u}^{v-1}y_iy_i^T-y_i\bar{y}_{u,v}^T-\bar{y}_{u,v}y_i^T+\bar{y}_{u,v}\bar{y}_{u,v}^T$$  
$$=\frac{1}{v-u}\sum_{i=u}^{v-1}y_iy_i^T-\frac{1}{v-u}\sum_{i=u}^{v-1}y_i\bar{y}_{u,v}^T-\frac{1}{v-u}\sum_{i=u}^{v-1}\bar{y}_{u,v}y_i^T+\frac{1}{v-u}\sum_{i=u}^{v-1}\bar{y}_{u,v}\bar{y}_{u,v}^T$$
$$=\frac{1}{v-u}\sum_{i=u}^{v-1}y_iy_i^T-\bar{y}_{u,v}\bar{y}_{u,v}^T-\bar{y}_{u,v}\bar{y}_{u,v}^T+\frac{v-u}{v-u}\bar{y}_{u,v}\bar{y}_{u,v}^T$$
$$=\frac{1}{v-u}\sum_{i=u}^{v-1}y_iy_i^T-\bar{y}_{u,v}\bar{y}_{u,v}^T$$  

$\bar{y}_{u,v}$ is computed from the difference of the cumulative sums: $$\bar{y}_{u,v}=\frac{1}{v-u}\sum_{i=u}^{v-1}y_i=\frac{1}{v-u}(\sum_{i=1}^{v-1}y_i-\sum_{i=1}^{u-1}y_i)$$  
And the quantity $\frac{1}{v-u}\sum_{i=u}^{v-1}y_iy_i^T$ is also a difference of cumulative sums:
$$\frac{1}{v-u}\sum_{i=u}^{v-1}y_iy_i^T=\frac{1}{v-u}(\sum_{i=1}^{v-1}y_iy_i^T-\sum_{i=1}^{u-1}y_iy_i^T)$$  
This method allows a faster computation of the covariance matrix in the different used algorithms.

## Feature selection based on correlation

For this method, features are selected based on their correlation. Features are considered highly correlated of the correlation is superior to the value of "seuil". For example, if a feature A is highly correlated to B and B is highly correlated to C, Only one of these features is kept. 

```python
def corr_selection(df,seuil):
    corr=df.corr()
    test=corr>seuil
    groups=dict()
    grouped=dict()
    group=0
    for i in test.columns:
        if (i not in grouped):
            grouped[i]=group
            groups[group]=[i]
            group=group+1
        
        for j in test.columns:
            
            if (i!=j)and(test[i][j]==True)and(j not in grouped):
                grouped[j]=grouped[i]
                groups[grouped[i]].append(j)
            

            elif (i!=j)and(test[i][j]==True)and(j in grouped)and(len(groups[grouped[i]])==1):
                del groups[grouped[i]]
                grouped[i]=grouped[j]
                
                groups[grouped[j]].append(i)
    select=[]
    for i in groups:
        select.append(groups[i][0])

    return df[select]
results1=score_illness("C:\\Users\\Yassine\\Desktop\\GaitData",rpt.BottomUp,MyCost(),'right',230,15,corr_sel=True,seuil=0.5)
```

## Steps identification using K-means.
For the previous methods, a covariance criterion was used to determine if a segment is a step or not. For this method, we cluster the steps, and then we identify which cluster is related to the steps.  
Since It is unknow which class corresponds to steps, 2 lists are constructed, one with the segments having the same class as the first segment, the other with segments having different classes. Then we return the list of segments having the highest f_score, which is most likely to correspond to the steps class.  
To each segment, we associate a vector of value which are the norm values of the describe method on each segment.  

Note: Agglomerative clustering gives nearly the same result as Kmeans. 


```python
from main import *
def process_data(link,method,cost,side,corr_sel=False,seuil=1):
    print("initiating process for: ",link)
    data=get_side_data(data_acquisition(link),side)
    meta_data=data[1]
    df=data[0]
    if (corr_sel==True):
        df=corr_selection(df,seuil)

    signals=np.asarray(df)
    steps=meta_data
    bkps=np.asarray(steps)
    bkps=bkps.reshape(bkps.shape[1]*bkps.shape[0])
    algo = method(custom_cost=cost).fit(signals)
    my_bkps = np.asarray(algo.predict(n_bkps=len(bkps)+1))[0:]
    segs=[]
    detected_steps=[]
    for i in range(len(my_bkps)-1):
        segs.append((my_bkps[i],my_bkps[i+1]))
    L=[]
    for s in segs:
        chunk=df.iloc[s[0]:s[1],:]
        scaled=chunk
        #scaled=pd.DataFrame(preprocessing.scale(chunk))
        summ=scaled.describe().iloc[1:8,]
    
        vector=np.linalg.norm(summ,axis=1)
        L.append(vector)
    L=np.asarray(L)
    kmeans = KMeans(n_clusters=2, random_state=0).fit(L)
    label_init=kmeans.labels_[0]
    dup=[]
    for i in range(len(segs)):
        if (kmeans.labels_[i]!=label_init):
            detected_steps.append(segs[i])
        else:
            dup.append(segs[i])
    p1=precision(detected_steps,steps)
    p2=precision(dup,steps)
    r1=recall(detected_steps,steps)
    r2=recall(dup,steps)
    f1=0
    f2=0
    if (p1+r1!=0):
        f1=f_score(detected_steps,steps)
    if (p2+r2!=0):
        f2=f_score(dup,steps)
    #print(f_score(dup,steps),"  ",f_score(detected_steps,steps))
    if (f1>f2):
        return (f1,detected_steps)
    else:
        return (f2,dup)
    
results2=score_illness("C:\\Users\\Yassine\\Desktop\\GaitData",rpt.BottomUp,MyCost(),'right',230,20)
```

```python
from main import *
def process_data(link,method,cost,side,corr_sel=False,seuil=1,PCA_t=False):
    print("initiating process for: ",link)
    data=get_side_data(data_acquisition(link),side)
    meta_data=data[1]
    df=data[0]
    data=df
    if (corr_sel==True):
        df=corr_selection(df,seuil)
    if (PCA_t==True):
        data=pd.DataFrame(preprocessing.scale(df))
        test=PCA()
        res=test.fit(data)
        var=np.cumsum(res.explained_variance_ratio_)<0.95
        i=[*var].index(False)
        PCA1=PCA(i)
        res=PCA1.fit(data)
        #comp=res.components_
        #var=np.cumsum(res.explained_variance_ratio_)
        #inert=res.singular_values_
        data=pd.DataFrame(res.transform(data))
    

    signals=np.asarray(data)
    steps=meta_data
    bkps=np.asarray(steps)
    bkps=bkps.reshape(bkps.shape[1]*bkps.shape[0])
    algo = method(custom_cost=cost).fit(signals)
    my_bkps = np.asarray(algo.predict(n_bkps=len(bkps)+1))[0:]
    segs=[]
    detected_steps=[]
    for i in range(len(my_bkps)-1):
        segs.append((my_bkps[i],my_bkps[i+1]))
    L=[]
    for s in segs:
        chunk=data.iloc[s[0]:s[1],:]
        scaled=chunk
        #scaled=pd.DataFrame(preprocessing.scale(chunk))
        summ=scaled.describe().iloc[1:8,]
    
        vector=np.linalg.norm(summ,axis=1)
        L.append(vector)
    L=np.asarray(L)
    kmeans = KMeans(n_clusters=2, random_state=0).fit(L)
    label_init=kmeans.labels_[0]
    dup=[]
    for i in range(len(segs)):
        if (kmeans.labels_[i]!=label_init):
            detected_steps.append(segs[i])
        else:
            dup.append(segs[i])
    p1=precision(detected_steps,steps)
    p2=precision(dup,steps)
    r1=recall(detected_steps,steps)
    r2=recall(dup,steps)
    f1=0
    f2=0
    if (p1+r1!=0):
        f1=f_score(detected_steps,steps)
    if (p2+r2!=0):
        f2=f_score(dup,steps)
    #print(f_score(dup,steps),"  ",f_score(detected_steps,steps))
    if (f1>f2):
        return (f1,detected_steps)
    else:
        return (f2,dup)
    
results_PCA=score_illness("C:\\Users\\Yassine\\Desktop\\GaitData",rpt.BottomUp,MyCost(),'right',230,20,PCA_t=True)
```
